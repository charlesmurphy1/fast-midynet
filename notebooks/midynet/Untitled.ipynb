```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
from scipy.special import loggamma
```

```python
def logfac(n):
    return loggamma(n + 1)

def logbinom(n, k):
    return logfac(n) - logfac(n - k) - logfac(k)

def binomial_pmf(k, n, p):
    return binom.pmf(k, n -1 , p)

def microbinom_pmf(k, n, m):
    N = n * (n - 1) / 2
    return np.exp(logbinom(n-1,k) + logbinom(N - n + 1, m - k) - logbinom(N, m))
```

```python
N, m = 1000, 2500
p = m / (N * (N - 1) / 2)
k = np.arange(100)

pmf1 = binomial_pmf(k, N, p)
pmf2 = microbinom_pmf(k, N, m)

fig, ax = plt.subplots(1, 1)
ax.plot(k, pmf1)
ax.plot(k, pmf2)
axx = ax.twinx()
axx.plot(k, (pmf1 - pmf2))
plt.show()
```
